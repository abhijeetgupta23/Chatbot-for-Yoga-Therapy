```python
import pandas as pd
```

```python
text_data = pd.concat([pd.read_csv("faq_data.csv"),pd.read_csv("blog_data.csv")])
```

```python
text_data
```

```python
q_a_list = []
```

```python
for index,row in text_data.iterrows():
    text_data_dict = {
        "Question": row['Question'],
        "Answer": row['Answer']
    }
    q_a_list.append(text_data_dict)
```

```python
q_a_list
```

```python
import marqo

mq = marqo.Client(url='http://localhost:8882')

mq.create_index("yoga_index_1")

mq.index("yoga_index").add_documents(q_a_list,
    tensor_fields=["Question"]
)

```

```python
results = mq.index("yoga_index_1").search(
    q="What is yoga therapy?"
)
```
