## Blog data

```python
import requests
from lxml import html
import time

# Fetch the page containing links
url = 'https://www.myyoganetwork.com/resources/blogs'
response = requests.get(url)
html_content = response.content

# Parse HTML content
tree = html.fromstring(html_content)

questions = []
answers = []

# Extract links using XPath
links = tree.xpath("/html/body/div[1]/section/div[2]/div/div[*]/div/div/div/h3/a/@href")

# Iterate through links, fetch content, and extract data
for link in links:
    # Construct full URL if the link is relative
    if not link.startswith('http'):
        link = url + link
    
    # Fetch content of each individual page
    page_response = requests.get(link)
    page_html_content = page_response.content
    
    # Parse HTML of individual page
    page_tree = html.fromstring(page_html_content)
    
    # Extract data from individual page using XPath
    title = page_tree.xpath("//h1/text()")[0]
    questions.append(title)

    article_content = page_tree.xpath("//div[@class='post-content']")[0].text_content()
    answers.append(article_content)
    
    # Add a delay to be respectful to the server
    time.sleep(1)  # Delay of 1 second between requests
```

```python
answers
```

```python
blog_data = pd.DataFrame({'Question':questions,'Answer':answers})
blog_data
```

```python
# Remove unwanted characters (replace '\r\n' with the characters you want to remove)
blog_data['Question'] = blog_data['Question'].str.replace('\r\n', '')
blog_data['Answer'] = blog_data['Answer'].str.replace('\r\n', '')
# Function to add space before capital letters
def add_space(text):
  return ''.join([' ' + char if char.isupper() else char for char in text])

# Apply the function to both columns
blog_data['Question'] = blog_data['Question'].apply(add_space)
blog_data['Answer'] = blog_data['Answer'].apply(add_space)
blog_data.head(10)
```

```python
blog_data.to_csv("blog_data.csv")
```

## FAQs

```python
import requests
from lxml import html
import time

# Fetch the page containing links
url = 'https://www.myyoganetwork.com/resources/faqs'
response = requests.get(url)
html_content = response.content

# Parse HTML content
tree = html.fromstring(html_content)
    
# Extract data from individual page using XPath
faq_question = tree.xpath("/html/body/div[1]/section/div[2]/div/div/div[*]/div/h5/text()")
print(faq_question)

faq_answer = tree.xpath("/html/body/div[1]/section/div[2]/div/div/div[*]/p/text()")
print(faq_answer)
```

```python
faq_data = pd.DataFrame({'Question':faq_question,'Answer':faq_answer})
faq_data
```

```python
faq_data.to_csv("faq_data.csv")
```
